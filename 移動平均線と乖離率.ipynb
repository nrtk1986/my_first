```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "colab"
import plotly.offline as offline
```

## **データの読み込み**

```
df = pd.read_csv('/content/drive/My Drive/SoftBank_20201104.csv', index_col = 0)
#期間の指定
df = df[df['Date'] >= '2019-01-01']
df.head(3)
```

## **移動平均の追加**

```
df['Move_25'] = df['Close'].rolling(window=25).mean()
df['Move_75'] = df['Close'].rolling(window=75).mean()
```

## **乖離率の追加**

```
df['PerOfDiv'] = (df['Close'] - df['Move_25']) * 100.0 / df['Move_25']
```

```
df.tail()
```

```
fig = go.Figure(
          data= [go.Candlestick(
                    x = df['Date'], 
                    open=df['Open'],
                    high=df['High'],
                    low=df['Low'],
                    close=df['Close'],
                    name = 'ローソク足'),
                 go.Scatter(
                    x = df['Date'],
                    y = df['Move_25'],
                    name = '25日移動平均',
                    yaxis = 'y1'),
                 go.Scatter(
                    x = df['Date'],
                    y = df['Move_75'],
                    name = '75日移動平均',
                    yaxis = 'y1'),
                 go.Scatter(
                    x = df['Date'],
                    y = df['PerOfDiv'],
                    name = '乖離率',
                    yaxis = 'y2')
                ],
          layout = go.Layout(
                      xaxis = dict(ticktext = df['Date']),
                      yaxis = dict(title = '株価', side = 'left', range = [min(df['Close']*0.8), max(df['Close']) * 1.2],showgrid=False),
                      yaxis2 = dict(title = '乖離率', side = 'right',overlaying = "y",  range = [min(df['PerOfDiv']*0.8), max(df['PerOfDiv']) * 1.2],showgrid=False)
                      )
          )
fig.show()
```
