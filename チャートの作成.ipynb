```
import pandas as pd
import pandas_datareader.data as web

import numpy as np
import plotly.graph_objects as go

import plotly.io as pio
pio.renderers.default = "colab"
import plotly.offline as offline
```

## 株価の取得

```
stockdata = web.DataReader("9984.JP", "stooq").dropna()
```

```
#index(Date)でソートして、Dateをindexからカラムに変更する
df = stockdata.sort_index().reset_index()
df.head()
```

## **株価の描画**

```
df['Move_25'] = df['Close'].rolling(window=25).mean()
df['Move_75'] = df['Close'].rolling(window=75).mean()
```

```
fig = go.Figure(
          data= [go.Candlestick(
                    x = df['Date'], 
                    open=df['Open'],
                    high=df['High'],
                    low=df['Low'],
                    close=df['Close'],
                    name = 'ローソク足'),
                 go.Scatter(
                    x = df['Date'],
                    y = df['Move_25'],
                    name = '25日移動平均'),
                 go.Scatter(
                    x = df['Date'],
                    y = df['Move_75'],
                    name = '75日移動平均')
                ],
          layout = go.Layout(
                      xaxis = dict(
                                ticktext = df['Date']
                              )
                      )
          )
fig.show()
```

```
#グラフの保存
offline.plot(fig, filename='sample.html', image="png")
```
